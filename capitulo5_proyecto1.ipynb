```python
#Capitulo 5.  Projectos Sencillos.

#Proyecto 1: Crear tu Primer Circuito Cu√°ntico


from qiskit import QuantumCircuit
from qiskit.providers.basic_provider import BasicProvider
from qiskit.visualization import plot_histogram

# Crear un circuito cu√°ntico con 1 qubit y 1 bit cl√°sico
circuit = QuantumCircuit(1, 1)

# Aplicar una puerta Hadamard para poner el qubit en superposici√≥n
circuit.h(0)

# Medir el qubit
circuit.measure(0, 0)

# Ejecutar el circuito en un simulador cu√°ntico
provider = BasicProvider()
backend = provider.get_backend("basic_simulator")
job = backend.run(circuit, shots=1024)

# Obtener los resultados
result = job.result()
counts = result.get_counts()

# Mostrar los resultados
print(f"Resultados de la medici√≥n: {counts}")

# Dibujar el circuito
circuit.draw(output='mpl')

# Visualizar los resultados en un histograma
plot_histogram(counts)

```
