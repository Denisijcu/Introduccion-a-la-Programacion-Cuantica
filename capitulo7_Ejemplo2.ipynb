```python
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

# Crear un circuito con 1 qubit y 1 bit clásico
circuit = QuantumCircuit(1, 1)

# Aplicar Hadamard para poner el qubit en superposición
circuit.h(0)

# Simular un error en el qubit
circuit.x(0)  # Error en el estado del qubit

# Medir el qubit
circuit.measure(0, 0)

# Ejecutar el circuito
sampler = Sampler()
job = sampler.run(circuit, shots=1024)
result = job.result()

# Mostrar los resultados
counts = result.quasi_dists[0].binary_probabilities()
print(f"Resultados con error cuántico: {counts}")

# Visualizar los resultados
plot_histogram(counts)

```
