```python
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

# Crear un circuito cuántico para simular sistemas caóticos
circuit = QuantumCircuit(3, 3)

# Aplicar Hadamard para poner los qubits en superposición
circuit.h([0, 1, 2])

# Aplicar puertas controladas para simular la dinámica del sistema
circuit.ccx(0, 1, 2)

# Medir los qubits
circuit.measure([0, 1, 2], [0, 1, 2])

# Ejecutar en un simulador
sampler = Sampler()
job = sampler.run(circuit, shots=1024)
result = job.result()

# Mostrar los resultados
counts = result.quasi_dists[0].binary_probabilities()
print(f"Resultados de la simulación del sistema: {counts}")

# Visualizar los resultados
plot_histogram(counts)
```
