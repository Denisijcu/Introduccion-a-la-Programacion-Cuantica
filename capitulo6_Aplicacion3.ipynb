```python
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler
from qiskit.visualization import plot_histogram

# Crear un circuito cuántico para el Algoritmo de Grover
circuit = QuantumCircuit(2, 2)

# Aplicar Hadamard a los qubits
circuit.h([0, 1])

# Aplicar la puerta CZ para marcar la solución
circuit.cz(0, 1)

# Aplicar Hadamard nuevamente
circuit.h([0, 1])

# Medir los qubits
circuit.measure([0, 1], [0, 1])

# Ejecutar en un simulador
sampler = Sampler()
job = sampler.run(circuit, shots=1024)
result = job.result()

# Mostrar los resultados
counts = result.quasi_dists[0].binary_probabilities()
print(f"Resultados del Algoritmo de Grover: {counts}")

# Dibujar el circuito
circuit.draw(output='mpl')

# Visualizar los resultados en un histograma
plot_histogram(counts)
```
