```python

# Importamos Qiskit
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Creamos un circuito cuántico con 2 qubits y 2 bits clásicos
circuit = QuantumCircuit(2, 2)

# Aplicamos una puerta Hadamard para colocar los qubits en superposición
circuit.h([0, 1])

# Medimos los qubits
circuit.measure([0, 1], [0, 1])

# Creamos una instancia del simulador
simulator = AerSimulator()

# Compilamos y ejecutamos el circuito
compiled_circuit = transpile(circuit, simulator)
result = simulator.run(compiled_circuit, shots=1024).result()

# Obtenemos los resultados y los mostramos
counts = result.get_counts(compiled_circuit)
print("Resultados:", counts)
plot_histogram(counts)
```
