```python
# Importamos las bibliotecas necesarias de Qiskit
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Creamos un circuito cu치ntico con un qubit
circuit = QuantumCircuit(1, 1)

# Aplicamos la puerta Hadamard al qubit para ponerlo en superposici칩n
circuit.h(0)

# Medimos el qubit y almacenamos el resultado
circuit.measure(0, 0)

# Dibujamos el circuito para visualizarlo
circuit.draw(output='mpl')

# Creamos una instancia del simulador
simulator = AerSimulator()

# Compilamos el circuito para el simulador
compiled_circuit = transpile(circuit, simulator)

# Ejecutamos el circuito en el simulador cu치ntico
job = simulator.run(compiled_circuit, shots=1024)
result = job.result()

# Obtenemos los resultados de las mediciones
counts = result.get_counts(compiled_circuit)
# Imprimimos los resultados
print("Resultados:", counts)

# Mostramos los resultados en un histograma
plot_histogram(counts)
```
