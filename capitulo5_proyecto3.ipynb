```python
from qiskit import QuantumCircuit
from qiskit.providers.basic_provider import BasicProvider
from qiskit.visualization import plot_histogram

# Crear un circuito de 2 qubits y 2 bits cl치sicos
circuit = QuantumCircuit(2, 2)

# Aplicar puertas Hadamard a ambos qubits
circuit.h([0, 1])

# Aplicar la puerta CZ (operador de Oracle) para marcar la soluci칩n
circuit.cz(0, 1)

# Aplicar Hadamard nuevamente a ambos qubits
circuit.h([0, 1])

# Medir ambos qubits
circuit.measure([0, 1], [0, 1])

# Configurar el proveedor y el backend para la simulaci칩n
provider = BasicProvider()
backend = provider.get_backend("basic_simulator")

# Ejecutar el circuito en el simulador cu치ntico
job = backend.run(circuit, shots=1024)

# Obtener los resultados
result = job.result()
counts = result.get_counts()

# Mostrar los resultados
print(f"Resultados del Algoritmo de Grover: {counts}")

# Dibujar el circuito
circuit.draw(output='mpl')

# Visualizar los resultados en un histograma
plot_histogram(counts)
```
